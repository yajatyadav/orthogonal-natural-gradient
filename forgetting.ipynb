```python
def compute_fgt(data):
    """
    Given a TxT data matrix, compute average forgetting at T-th task
    """
    num_tasks = data.shape[0]
    T = num_tasks - 1
    fgt = 0.0
    task_fgt = []
    for i in range(T):
        x = np.max(data[:T, i])
        y = data[T, i]
        print(x, y, x-y)
        fgt_i = x - y
        task_fgt.append(fgt_i)
        fgt += fgt_i
    avg_fgt = fgt/ float(num_tasks - 1)
    return avg_fgt, task_fgt
```
